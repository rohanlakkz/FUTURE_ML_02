```python
import pandas as pd

df = pd.read_csv("/content/customer_support_tickets.csv")

df.head()
```

```python
#checking coulumn names
print(df.columns)
```

```python
df.head(10)
```

```python
#checking missing values
df.isnull().sum()
```

```python
#drop missing values
df = df.dropna()
```

```python
import nltk
import re
import string

nltk.download('stopwords')
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

def clean_text(text):

    text = text.lower()

    text = re.sub(r'\d+', '', text)

    text = text.translate(str.maketrans('', '', string.punctuation))

    words = text.split()

    words = [w for w in words if w not in stop_words]

    return " ".join(words)
```

```python
#creating text cleaning function
df['clean_text'] = df['Ticket Description'].apply(clean_text)

df[['Ticket Description','clean_text']].head()
```

```python
print(df.columns)
```

```python
#convert text to numbers
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()

X = vectorizer.fit_transform(df['clean_text'])

print(X.shape)
```

```python
#preparing categorical ticket types
y_category = df['Ticket Type']

print(y_category.value_counts())
```

```python
#spliting data for training and testing
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y_category,
    test_size=0.2,
    random_state=42
)

print(X_train.shape)
print(X_test.shape)
```

```python
#training category classsification model
from sklearn.linear_model import LogisticRegression

model_cat = LogisticRegression()

model_cat.fit(X_train, y_train)
```

```python
#predicting ticket categories
pred_cat = model_cat.predict(X_test)

pred_cat[:10]
```

```python
#evaluating categorical model
from sklearn.metrics import accuracy_score, classification_report

print("Accuracy:", accuracy_score(y_test, pred_cat))

print(classification_report(y_test, pred_cat))
```

```python
#preparing priority labels
y_priority = df['Ticket Priority']

print(y_priority.value_counts())
```

```python
#spliting data for priority prdiction
X_train2, X_test2, y_train2, y_test2 = train_test_split(
    X,
    y_priority,
    test_size=0.2,
    random_state=42
)

print(X_train2.shape)
print(X_test2.shape)
```

```python
#training priority prediction model
model_pri = LogisticRegression()

model_pri.fit(X_train2, y_train2)
```

```python
#predicting ticket priority
pred_pri = model_pri.predict(X_test2)

pred_pri[:10]
```

```python
#evaluating priority model
print("Accuracy:", accuracy_score(y_test2, pred_pri))

print(classification_report(y_test2, pred_pri))
```

```python
#testing demo
def predict_ticket(text):

    clean = clean_text(text)

    vector = vectorizer.transform([clean])

    category = model_cat.predict(vector)
    priority = model_pri.predict(vector)

    print("Predicted Category:", category[0])
    print("Predicted Priority:", priority[0])
```

```python
predict_ticket("My payment was deducted but order not confirmed")
```
